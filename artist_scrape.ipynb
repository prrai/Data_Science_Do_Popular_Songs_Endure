Code to scrape http://www.billboard.com/ for artists and get their number of songs which have topped charts, have been in top 10 and have been in top 100.

```python

from bs4 import BeautifulSoup
import pandas as pd
import urllib.request as ur
import urllib
import re


songs = pd.read_csv('January_5_1959.csv')
homepage = "http://www.billboard.com/"

songs['TopSongsArtist'] = ''
songs['TopSongsArtist10'] = ''
songs['TopSongsArtist100'] = ''

# site = ur.urlopen(homepage)

# artists = songs['artist']

for index, row in songs.iterrows():
    
    artist = row['Artist']
    
    top100 = 0
    top10 = 0
    top = 0
    
    artist = re.sub('\s', '-', artist)    
    page= homepage + "music/" + artist
    
    try:
       site = ur.urlopen(page)
    except:
        continue
#     site = ur.urlopen(page)
    
    soup = BeautifulSoup(site)
        
    spans = soup.find_all('span', {'class' : 'artist-section--chart-history__stats__stat'})
    
    spannums = soup.find_all('span', {'class' : 'artist-section--chart-history__stats__stat--number'})

    top = 0
    top10 = 0
    top100 = 0
    
    for span in spans:
        
        stats = span.text
        
        label = stats.split(' ', 1)[1]
        value = stats.split(' ', 1)[0]
        
        if label == "No. 1 Hits":
            top = value
        
        if label == "Top 10 Hits":
            top10 = value
            
        if label == "Songs":
            top100 = value
    
    
    songs.set_value(index, 'TopSongsArtist', top)
    songs.set_value(index, 'TopSongsArtist10', top10)
    songs.set_value(index, 'TopSongsArtist100', top100)


    print(artist)
    print (top)
    print (top10)
    print (top100)
    

songs.to_csv('outsongs.csv', index=False)
```
